```python
arrangement = "()(((()())(())()))(())"


def solution(arrangement):
    num =0 
    num_of_bar=0
    for i,value in enumerate(arrangement):
        if value == '(':
            num_of_bar +=1
        elif value == ')' and arrangement[i-1] == '(':
            num_of_bar -=1
            num += num_of_bar
        else:
            num_of_bar -= 1
            num +=1
    answer = num
    return answer
solution(arrangement)
```

```python
priorities = [1,1,9,1,1,1]
def solution(priorities, location):
    print_num =1
    
    while(location != -1):

        print("list :{} ".format(priorities))
        print("location: %d" %location)
        if priorities[0] == max(priorities) and location ==0:
            answer = print_num
            location = -1
        elif priorities[0] == max(priorities) and location !=0:
            print_num +=1
            print("print_num: %d"%print_num)
            del priorities[0]
            location -= 1
        elif priorities[0] != max(priorities) and location ==0:
            priorities.append(priorities[0])
            del priorities[0]
            location += (len(priorities)-1)
        else:
            priorities.append(priorities[0])
            del priorities[0]
            location -= 1
                
   
    return answer
solution(priorities,0)
```

```python
import numpy as np
bridge_length =100
weight = 100
truck_weights = [10]

def solution(bridge_length, weight, truck_weights):
    time =0
    bridge=list(np.zeros(bridge_length,dtype=int))
    merge_bridge = bridge+truck_weights
    while sum(merge_bridge) != 0:
        if sum(merge_bridge[:bridge_length+1]) < weight:
           del merge_bridge[0]
           merge_bridge.append(0)
        else:
           del merge_bridge[0]
           merge_bridge.insert(bridge_length-1,0)
        
        time += 1
        print(merge_bridge)
    return time
solution(bridge_length,weight,truck_weights)
```

```python
heights = [3,9,9,3,5,7,2]

def solution(heights):
    re_heights = heights[::-1]
    for i in range(len(re_heights)-1):
       
        for j in range(len(re_heights)):
            
            if j<=i:
                continue
            else:
                value = re_heights[i]
                re_heights[i] = len(re_heights)
                if value<re_heights[j]:
                    re_heights[i]=j
                    print(j)
                    
                    break 
    re_heights[i+1]=len(re_heights)
    for i in range(len(re_heights)):
        re_heights[i] = len(re_heights) - re_heights[i]
    answer = re_heights[::-1]
    return answer
solution(heights)
```
